```python
from TweetCrawler import *
from config import *
from models.ModelMarch import *
import matplotlib.pyplot as plt
from utils import *

"""
Source: https://matplotlib.org/stable/gallery/pie_and_polar_charts/pie_features.html
"""

### QUERY TWEETS
tweetCrawler = TweetCrawler(DATABASE_URI_RDS_TWEETS)
query_df = tweetCrawler.crawl_data_with_session(March_Tweets, filtering='% march %')
#print("There are ", len(query_df), " tweets that include the word 'march'")

count_of_movement = count_tweets(query_df, 'Movement')
count_of_month = count_tweets(query_df, 'Month')

#print(count_of_movement)
#print(count_of_month)

### BULD MATPLOTLIB PIE CHART
labels = 'march is an \'Movement\'', 'march as in \'Month\''
sizes = [count_of_movement, count_of_month]
explode = (0.1, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')

ax1.set_title('The proportion of \'march\' meanings in the keyword filtered Bristol set (13,689 tweets)\n')

plt.show()
```
